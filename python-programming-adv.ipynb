```python
# question-1
class Solution:
    def runningSum(self, nums):
        for i in range(1, len(nums)):
            nums[i] += nums[i-1]
        return nums
s = Solution()
nums = list(map(int,input().split(",")))
print(nums)
print(s.runningSum(nums))
```

```python
#question-2
class Solution:
    def shuffle(self, nums, n):
        result = []
        for i in range(n):
            result.append(nums[i])
            result.append(nums[i + n])
        return result
nums = list(map(int, input().split(",")))
print(nums)
n = int(input())
print(n)
s = Solution()
print(s.shuffle(nums, n))
```

```python
#question-3:
class Solution:
    def kidsWithCandies(self, candies, extraCandies):
        max_candy = max(candies)
        result=[]
        for c in candies:
            result.append(c + extraCandies >= max_candy)
        return result
candies = list(map(int, input().split(",")))
print(candies)
extraCandies = int(input())
print(extraCandies)
s = Solution()
print(s.kidsWithCandies(candies,extraCandies))
```
