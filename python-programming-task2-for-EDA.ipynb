```python
# Defanging an IP Address
class Solution:
    def defangIPaddr(self, address: str) -> str:
        return address.replace(".", "[.]")
address = input("Enter IP address: ")
obj = Solution()
print("Defanged IP Address:", obj.defangIPaddr(address))
```

```python
#Find Numbers with Even Number of Digits
class Solution:
    def findNumbers(self, nums):
        count = 0
        for num in nums:
            if len(str(num)) % 2 == 0:
                count += 1
        return count
nums = list(map(int, input("Enter numbers separated by space: ").split()))
print(nums)
obj = Solution()
print("Count of numbers with even number of digits:", obj.findNumbers(nums))
```

```python
#Number of Good Pairs
class Solution:
    def numIdenticalPairs(self, nums):
        count = 0
        n = len(nums)

        for i in range(n):
            for j in range(i + 1, n):
                if nums[i] == nums[j]:
                    count += 1
        return count
nums = list(map(int, input("Enter numbers separated by space: ").split()))
print(nums)
obj = Solution()
print("Number of good pairs:", obj.numIdenticalPairs(nums))
```

```python
# How Many Numbers Are Smaller Than the Current Number
class Solution:
    def smallerNumbersThanCurrent(self, nums):
        result = []
        for i in range(len(nums)):
            count = 0
            for j in range(len(nums)):
                if nums[j] < nums[i]:
                    count += 1
            result.append(count)
        return result
nums = list(map(int, input("Enter numbers separated by space: ").split()))
print(nums)
obj = Solution()
print("Result:", obj.smallerNumbersThanCurrent(nums))
```

```python
#Subtract the Product and Sum of Digits of an Integer
class Solution:
    def subtractProductAndSum(self, n):
        product = 1
        summation = 0

        while n > 0:
            digit = n % 10
            product *= digit
            summation += digit
            n //= 10

        return product - summation
n = int(input("Enter an integer: "))
print(n)
obj = Solution()
print("Result:", obj.subtractProductAndSum(n))
```

```python
#  XOR Operation in an Array
class Solution:
    def xorOperation(self, n, start):
        result = 0
        for i in range(n):
            result ^= start + 2 * i
        return result
n = int(input("Enter value of n: "))
print(n)
start = int(input("Enter value of start: "))
print("start:",start)
obj = Solution()
print("XOR Result:", obj.xorOperation(n, start))
```
