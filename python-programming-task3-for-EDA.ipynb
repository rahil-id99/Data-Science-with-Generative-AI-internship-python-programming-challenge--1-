```python
#Maximum Product of Two Elements in an Array
class Solution:
    def maxProduct(self, nums):
        nums.sort()
        return (nums[-1] - 1) * (nums[-2] - 1)
nums = list(map(int, input("Enter numbers separated by space: ").split()))
print(nums)
obj = Solution()
print("Maximum Product:", obj.maxProduct(nums))
```

```python
#Count Number of Teams
class Solution:
    def numTeams(self, rating):
        n = len(rating)
        teams = 0

        for j in range(n):
            left_smaller = left_greater = 0
            right_smaller = right_greater = 0

            for i in range(j):
                if rating[i] < rating[j]:
                    left_smaller += 1
                elif rating[i] > rating[j]:
                    left_greater += 1

            for k in range(j + 1, n):
                if rating[k] < rating[j]:
                    right_smaller += 1
                elif rating[k] > rating[j]:
                    right_greater += 1

            teams += left_smaller * right_greater + left_greater * right_smaller

        return teams
rating = list(map(int, input("Enter ratings separated by space: ").split()))
print(rating)
obj = Solution()
print("Number of valid teams:", obj.numTeams(rating))
```

```python
#Number of Students Doing Homework at a Given Time
class Solution:
    def busyStudent(self, startTime, endTime, queryTime):
        count = 0
        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1
        return count
startTime = list(map(int, input("Enter start times separated by space: ").split()))
endTime = list(map(int, input("Enter end times separated by space: ").split()))
queryTime = int(input("Enter query time: "))
obj = Solution()
print("Number of students doing homework:", obj.busyStudent(startTime, endTime, queryTime))
```

```python
#Number of Steps to Reduce a Number to Zero
class Solution:
    def numberOfSteps(self, num):
        steps = 0
        while num > 0:
            if num % 2 == 0:
                num //= 2
            else:
                num -= 1
            steps += 1
        return steps
num = int(input("Enter a number: "))
print(num)
obj = Solution()
print("Number of steps:", obj.numberOfSteps(num))
```

```python
# Counting Bits
class Solution:
    def countBits(self, n):
        ans = [0] * (n + 1)
        for i in range(1, n + 1):
            ans[i] = ans[i // 2] + (i % 2)
        return ans
n = int(input("Enter a number n: "))
print(n)
obj = Solution()
print("Count of bits from 0 to n:", obj.countBits(n))
```
